```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
def count():

    landmarks_path = '/vagrant/imgs/list_attr_celeba.csv'
    df = pd.read_csv(landmarks_path)
    print(len(df))
    output = []
    for index, row in df.iterrows():
        label = np.array(row.values.tolist()[1:])
        label[label < 0] = 0
        output.append(np.sum(label))
                
    return np.array(output)
```

```python
output = count()
```

```python
counts = {}
for i in range(1, 21): # none over 20
    counts[i] = np.sum(output[output == i])
    
```

```python
n_groups = len(counts)

ones = tuple(counts.values())
# std_men = (2, 3, 4, 1, 2)


fig, ax = plt.subplots(figsize=(15,10))

index = np.arange(n_groups)
bar_width = 0.8

opacity = 0.4
error_config = {'ecolor': '0.3'}

rects1 = ax.bar(index, ones, bar_width,
                alpha=opacity, color='b')


ax.set_xlabel('Number of Attributes')
ax.set_ylabel('Number of Images')
ax.set_title('Attributes Per Image Counts')
ax.set_xticks(index + bar_width / 2)
ax.set_xticklabels(tuple(counts.keys()))
# ax.legend()

# fig.tight_layout()
plt.show()
```
