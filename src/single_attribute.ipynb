```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os
import math
import matplotlib.image as mpimg
```

```python
images_path = '/vagrant/imgs/images/img_align_celeba'
landmarks_path = '/vagrant/imgs/list_attr_celeba.csv'
```

```python
def count():
    df = pd.read_csv(landmarks_path)
    output = []
    for index, row in df.iterrows():
        label = np.array(row.values.tolist()[1:])
        label[label < 0] = 0
        output.append(np.sum(label))
                
    return np.array(output)
```

```python
def images(output, num):
    df = pd.read_csv(landmarks_path)
    rows = df.iloc[list(np.where(output==num)[0])]
    return rows['image_id'].values.tolist()
```

```python
def get_rows(image_list, cols=5):
    rows = math.ceil(len(image_list)/cols)
    return cols, rows
```

```python
def get_headers():
    df = pd.read_csv(landmarks_path)
    headers = list(df)
    return headers[1:]
```

```python
def get_attribute(image_path):
    df = pd.read_csv(landmarks_path)
    row = df.loc[df['image_id'] == image_path]
    label = np.array(row.values.tolist()[0][1:])
    return np.where(label == 1)[0][0]
```
