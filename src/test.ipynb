```python
def resize_point(old_x, old_y, old_shape, new_shape):
    _x = new_shape[0]
    _i = old_shape[0]
    print('{} * {} / {}'.format(old_x, _x, _i))
    new_x = old_x * (float(new_shape[0]) / float(old_shape[0]))
    new_y = old_y * (float(new_shape[1]) / float(old_shape[1]))
    return int(new_x), int(new_y)

```

```python

import pandas as pd

def get_datum(row, label):
    return row.get(label).values[0]

def all_land(row):
    lefteye = get_datum(row, 'lefteye_x'), get_datum(row, 'lefteye_y')
    righteye = get_datum(row, 'righteye_x'), get_datum(row, 'righteye_y')
    nose = get_datum(row, 'nose_x'), get_datum(row, 'nose_y')
    leftmouth = get_datum(row, 'leftmouth_x'), get_datum(row, 'leftmouth_y')
    rightmouth = get_datum(row, 'rightmouth_x'), get_datum(row, 'rightmouth_y')
    return lefteye, righteye, nose, leftmouth, rightmouth


path = '/vagrant/imgs/small/images'
landmarks_path = '/vagrant/imgs/small/list_landmarks_align_celeba.csv'
directory = os.listdir(path)
df = pd.read_csv(landmarks_path)

for f in directory:
    image = cv2.imread(os.path.join(path, f))
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    row = df.loc[df['image_id'] == f]
    landmarks = all_land(row)
    print(landmarks)
    print(image.dtype)
    for center in landmarks:
        cv2.circle(image, center, 2, color=(0,0,255), thickness=-1)
    plt.imshow(image)
    plt.title('my picture')
    plt.show()
    
    
    
```

```python
def get_bbox(row):
    x = get_datum(row, 'x_1')
    y = get_datum(row, 'y_1')
    width = get_datum(row, 'width')
    height = get_datum(row, 'height')
    return (x,y),(x-width,y+height)

path = '/vagrant/imgs/orig_images/small'
bbox_path = '/vagrant/imgs/small/list_bbox_celeba.csv'
directory = os.listdir(path)
df = pd.read_csv(bbox_path)
#cv2.rectangle(img, pt1, pt2, color[, thickness[, lineType[, shift]]])  None
for f in directory:
    image = cv2.imread(os.path.join(path, f))
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    row = df.loc[df['image_id'] == f]
    print(row)
    pt1, pt2 = get_bbox(row)
    print(pt1, pt2)
    cv2.rectangle(image, pt1, pt2, color=(0,0,255))
    plt.imshow(image)
    plt.title('my picture')
    plt.show()
```
